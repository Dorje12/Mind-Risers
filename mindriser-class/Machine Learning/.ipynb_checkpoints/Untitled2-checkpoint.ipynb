```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
import statsmodels.api as sm
import seaborn as sns
```

```python
people = pd.read_csv("income.data.csv")
people.head()
```

```python
people_final = people.dropna()
people_final.reset_index(inplace=True, drop = True)
```

```python
people_final
```

```python
sns.pairplot(people_final)
```

```python
people_final.columns
```

```python
old_data = people_final[["income","happiness"]]
```

```python
old_formula='happiness ~ income'
```

```python
from statsmodels.formula.api import ols
```

```python
OLS = ols(formula =old_formula, data = old_data)
model =OLS.fit()
```

```python
model.summary()
```

```python
sns.regplot(x='income', y='happiness', data =old_data )
```
