```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
people = pd.read_csv("income.data.csv")
people.head()
```

```python
people_final = people.dropna()
people_final.reset_index(inplace=True, drop = True)
```

```python
income = people_final["income"].to_numpy()
income
```

```python
happiness = people_final["happiness"].to_numpy()
happiness
```

```python
plt.figure(figsize=(10,8))
plt.scatter(income, happiness, color='cyan')
plt.title('income vs happiness')
plt.xlabel('income')
plt.ylabel('happiness')
plt.grid(True)
plt.show()
```

```python
plt.hist(people_final)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(income.reshape(-1,1), happiness, test_size=0.2, random_state=42)

print(f'X_train shape: {X_train.shape}, y_train shape: {y_train.shape}')
print(f'X_test shape: {X_test.shape}, y_test shape: {y_test.shape}')
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
mse = mean_squared_error(y_test,y_pred)
rmse = np.sqrt(mse)
print(f'Room Mean Squared Error (RMSE): {rmse}')
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(X_test, y_test, color='blue')
plt.plot(X_test, y_pred, color='cyan', linewidth=2)
plt.title('Human happiness index predecton')
plt.xlabel('income')
plt.ylabel('happiness')
plt.grid(True)
plt.show()
```
