```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
import statsmodels.api as sm
```

```python
education_districtwise = pd.read_csv("education_districtwise.csv")
education_districtwise = education_districtwise.dropna()
```

```python
sampled_data = education_districtwise.sample(n=50, replace= False, random_state= 31208)
sampled_data
```

```python
sample_mean = sampled_data['OVERALL_LI'].mean()
sample_mean
```

```python
estimated_standard_error = sampled_data['OVERALL_LI'].std()/np.sqrt(sampled_data.shape[0])
estimated_standard_error
```

```python
lower_bound, upper_bound = stats.norm.ppf([0.025, 0.975], loc=sample_mean, scale=estimated_standard_error)
```

```python
print(f'({lower_bound},{upper_bound})')
```

```python
lower_bound, upper_bound = stats.norm.ppf([0.005, 0.995], loc=sample_mean, scale=estimated_standard_error)
```

```python
print(f'({lower_bound},{upper_bound})')
```
