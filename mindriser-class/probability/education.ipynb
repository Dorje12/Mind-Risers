```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
```

```python
education_districtwise = pd.read_csv("education_districtwise.csv")
education_districtwise = education_districtwise.dropna()
```

```python
education_districtwise['OVERALL_LI'].hist()
```

```python
education_districtwise.describe()
```

```python
mean_overall_li = education_districtwise['OVERALL_LI'].mean()
mean_overall_li
```

```python
std_overall_li = education_districtwise['OVERALL_LI'].std()
std_overall_li
```

```python
lower_limmit1 = mean_overall_li-1* std_overall_li
upper_limit1 = mean_overall_li+1* std_overall_li

lower_limmit2 = mean_overall_li-2* std_overall_li
upper_limit2= mean_overall_li+2* std_overall_li

lower_limmit3 = mean_overall_li-3* std_overall_li
upper_limit3 = mean_overall_li+3* std_overall_li
```

```python
((education_districtwise["OVERALL_LI"] >= lower_limmit1) & (education_districtwise["OVERALL_LI"] <=upper_limit1)).mean()
```

```python
((education_districtwise["OVERALL_LI"] >= lower_limmit2) & (education_districtwise["OVERALL_LI"] <=upper_limit2)).mean()
```

```python
((education_districtwise["OVERALL_LI"] >= lower_limmit3) & (education_districtwise["OVERALL_LI"] <=upper_limit3)).mean()
```

```python
education_districtwise['Z_SCORE'] = stats.zscore(education_districtwise['OVERALL_LI'])
education_districtwise
```
